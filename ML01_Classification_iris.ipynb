```
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
# print(iris.DESCR)
# → print(iris.DESCR) 可顯示該資料集詳細描述
# print(iris)
# print(iris.feature_names) 
# → ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
# print(iris.target_names)
# → ['setosa' 'versicolor' 'virginica']
df = pd.DataFrame(iris["data"],
      columns=iris["feature_names"])
df["ans"] = iris["target"]
df
```

```
import numpy as np
from sklearn.model_selection import train_test_split
# [題目90, 題目10, 答案90, 答案10]
#x_train = np.array(iris["data"]) 90%
#x_test = np.array(iris["data"]) 10%
#y_train = np.array(iris["target"]) 90%
#y_test = np.array(iris["target"]) 10%
x_train, x_test, y_train, y_test = train_test_split(np.array(iris["data"]),
                             np.array(iris["target"]),
                             train_size = 0.9)
```

```
#建立分類器 max_depth = 2  分兩層
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2)
#clf.fit() 用訓練數據擬合分類器模型
clf.fit(x_train, y_train)
```

```
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
#plt.figure() 顯示大小
plt.figure(figsize=(15, 15))
#feature_names 分類原因 class_names分類名稱 filled 上色
plot_tree(clf, 
     feature_names=iris["feature_names"], 
     class_names=iris["target_names"], 
     filled=True)
```

```
from sklearn.metrics import accuracy_score
#clf.predict(x_test) 預測
pre = clf.predict(x_test)
accuracy_score(pre, y_test)
```
